```python
#Splitting and Printing the Folds

import pandas as pd
data = pd.read_csv('auto_mpg.csv')
len(data)
from sklearn.model_selection import train_test_split
data.shape
#1st Important Line
data_train, data_test = train_test_split(data, test_size=0.3, random_state=123)
len(data_train)
len(data_test)
#2nd Important Line
from sklearn.model_selection import KFold
kf=KFold(n_splits=10, shuffle=True, random_state=123)
#3rd Important Line
for fold, (train_index, test_index) in enumerate(kf.split(data),1):
    print(f"Fold{fold}")
    print("Train Indices:", train_index)
    print("Test Indices:",test_index,"\n")
```
