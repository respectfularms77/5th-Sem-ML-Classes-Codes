```python
#Sampling distribution variance vs sample size

population = np.random.normal(loc=5.0,scale=2.0,size=200000)
true_sigma2 = population.var()
sample_sizes = [2,5,10,20,50,100]
empirical_vars = []

for n in sample_sizes:
    sample_means = [np.mean(np.random.choice(population,n)) for _ in range(2000)]
    empirical_vars.append(np.var(sample_means))
    
for n,ev in zip(sample_sizes,empirical_vars):
    print(f'n={n}: empirical Var(mean)={ev:.5f},theoretical sigma^2/n={true_sigma2/n:.5f}')
          
#Plot
plt.figure(figsize=(7,4))
plt.plot(sample_sizes,empirical_vars,marker='o')
plt.plot(sample_sizes,[true_sigma2/n for n in sample_sizes],linestyle='dashdot')
plt.xlabel('Sample size n')
plt.ylabel('Variable of sample mean')
plt.title('Sampling distribution variance vs n')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd

from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
from sklearn.model_selection import train_test_split
df=pd.read_csv("btissue.csv")
print(df)
X=df.iloc[:, :-1].values
y=df.iloc[:, -1].values
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
knn = KNeighborsClassifier(n_neighbors=3)
model=knn.fit(X_train,y_train)
y_pred=model.predict(X_test)

print("\n---KNN Results ---")
print("Accuracy:",accuracy_score(y_test,y_pred))
print("Precision:",precision_score(y_test,y_pred,average='macro',zero_division=0))
print("Recall:",recall_score(y_test,y_pred,average='macro',zero_division=0))
print("F1-score:",f1_score(y_test,y_pred,average='macro',zero_division=0))

cm=confusion_matrix(y_test,y_pred)
plt.figure(figsize=(4,3))
sns.heatmap(cm,annot=True,fmt='d',cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
