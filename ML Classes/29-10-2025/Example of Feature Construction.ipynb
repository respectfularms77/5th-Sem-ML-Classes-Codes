```python
#Create a Dictinary having Keys - (City, Athelete, & Chance of Wins) having of 4 Values each

import pandas as pd
data = {
    'City' : ['Pune', 'Bhopal', 'Delhi', 'Assam'],
    'Athelete' : ['A', 'B', 'C', 'D'],
    'C_Wins' : [1,0,0,1]
}
df = pd.DataFrame(data)
display(df)
display(pd.get_dummies(df,drop_first=True))



data1 = {
    'Len' : [10, 15, 20, 20],
    'Bred' : [2, 3,  5, 6],
    'Price' : [1000,2000,3000,4000]
}
df1 = pd.DataFrame(data1)
#Feature Construction
df1['Area'] = df1['Len'] * df1['Bred']
display(df1)
display(pd.get_dummies(df1,drop_first=True))
```
