```python
import numpy as np
import pandas as pd

# Load CSV File
df = pd.read_csv("btissue.csv")
print("\n ---- DATASET PREVIEW -----")
display(df.head(10))

#Basic Info
print("\n ---- DATASET INFO -----")
display(df.info())

# Check for missing Value
print("\n Missing Values per coloumn:")
print(df.isnull().sum())

#Summary Statistics
print("\n\n ---- Summary Statistics-----")
print(df.describe())
```

```python
X = df.iloc[:, :-1].values      #Features
y = df.iloc[:,-1].values        #Target

print("\nFeatures Shape:",X.shape)
print("\nTarget's Shape:",y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print("\nTrain Shape:",X_train.shape)   #Number of rows and coloumns in the training features
print("\nTest Shape:",X_test.shape)     #Number of rows and coloumns in the testing features
```
