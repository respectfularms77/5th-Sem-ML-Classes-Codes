```python
"""Q1. Data Holdout:
Load the dataset auto-mpg.csv and perform the following tasks:
1. Print the shape of the dataset.
2. Split the dataset into training (80%) and testing (20%) subsets.
3. Use a random state for reproducibility.
4. Print the training and test sets and their length"""

import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv("auto_mpg.csv")
print("The Shape of the Data: ",data.shape)

train, test = train_test_split(data, test_size=0.2, random_state=42)

print("Training Dataset:\n",train)
print("Testing Dataset:\n",test)
print("Length of Training Dataset:\n",len(train))
print("Length of Testing Dataset:\n",len(test))
```

```python
"""Q2. Implementing 5-Fold Cross-Validation
Using the dataset btissue.csv, do the following:
1. Create a 5-fold cross-validation.
2. For each fold (starting with 0), print:
 - fold number
 - indices of the training set
 - indices of the test set
3. Print the training set and test set and their length for the 3rd iteration."""

import pandas as pd
from sklearn.model_selection import KFold

data = pd.read_csv('btissue.csv')

kf = KFold(n_splits=5, shuffle=True, random_state=42)

for fold, (train_indices, test_indices) in enumerate(kf.split(data)):
    print(f"\n\nFold {fold}:")
    print("Training set indices:", train_indices)
    print("Test set indices:", test_indices)
    # for the 3rd iteration (fold 2)
    if fold == 2:
        train_data = data.iloc[train_indices]
        test_data = data.iloc[test_indices]
        print("\nTraining set for Fold 2:\n", train_data)
        print("\nTest set for Fold 2:\n", test_data)
        print("\nLength of Training set for Fold 2:", len(train_data))
        print("Length of Test set for Fold 2:", len(test_data),"\n")
```

```python
"""Q3. Bootstrap Sampling
With the dataset btissue.csv, perform the following:
1. Extract the predictors.
2. Generate a bootstrap sample of size 100.
3. Display the first 10 rows of the resampled data."""

import pandas as pd

data = pd.read_csv("btissue.csv")

# Extracting predictors (all columns except the last one which is the target variable)
predictors = data.iloc[:, :-1]

# Generate a bootstrap sample of size 100
# The replace=True argument is crucial for bootstrap sampling
bootstrap_sample = predictors.sample(n=100, replace=True, random_state=42)

# Displaying the first 10 rows of the resampled data
print("First 10 rows of the bootstrap sample:")
display(bootstrap_sample.head(10))
```

```python
"""Q4. Random Forest Classification and Evaluation
Using the dataset btissue.csv, carry out the steps below:
1. Separate predictors and target.
2. Perform a holdout split (80% training, 20% testing).
3. Train a Random Forest Classifier.
4. Evaluate the model using:
 - accuracy_score
 - confusion_matrix"""

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
# import joblib     # Import joblib to save the trained model

data = pd.read_csv('btissue.csv')

# 1. Separate predictors and target
predictors = data.iloc[:, :-1]  # All columns except the last one
target = data.iloc[:, -1]     # The last column

# 2. Perform a holdout split (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.2, random_state=90)

# 3. Train a Random Forest Classifier
rf = RandomForestClassifier(random_state=123)
rf.fit(X_train, y_train)

# 4. Evaluate the model
y_predicted = rf.predict(X_test)

print("Accuracy Score:", accuracy_score(y_test, y_predicted))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_predicted))

# 5. Save the trained model
# joblib.dump(rf, 'random_forest_model.pkl')
# print("\nTrained model saved as 'random_forest_model.pkl'")
```

```python
"""Q5. Regression and Evaluation
Load the dataset auto_mpg.csv and perform the following tasks:
1. Remove all rows where value of any column is ‘NaN’.
2. Separate predictors and target.
3. Perform a holdout split (80% training, 20% testing).
4. Train a Linear Regression Model.
5. Evaluate the model using:
 - mean_squared_error
 - r2_score"""

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

data = pd.read_csv('auto_mpg.csv')

data = data.dropna(axis=0, how='any')
predictors = data.iloc[:,1:8]
target = data.iloc[:,0]
X_train, X_test, Y_train, Y_test = train_test_split(predictors, target, test_size = 0.25, random_state = 90)
model = LinearRegression().fit(X_train, Y_train)
Y_predicted = model.predict(X_test)
print("Mean Squared Error:",mean_squared_error(Y_test,Y_predicted))
print(r2_score(Y_test, Y_predicted))
```

```python
"""Q6. Clustering and Quality Measures
Using the dataset Dataset_spine.csv:
1. Exclude the class (last) column to form predictor variables.
2. Apply K-Means clustering with n_clusters=3 and random_state=123.
3. Compute and print:
 - v-measure score
 - Silhouette score"""

import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics.cluster import v_measure_score
from sklearn.metrics import silhouette_score

data = pd.read_csv("Dataset_spine.csv")
# print(data)
data1 = data.iloc[:,:11]
# print(data1) # Kept commented as per original
data2 = data.iloc[:,12]

# FIX: Added n_init=10 to explicitly set the initialization runs
kmeans = KMeans(n_clusters=3, random_state=123, n_init="auto")
kmeans.fit(data1)
cluster_labels = kmeans.predict(data1)

v = v_measure_score(data2, cluster_labels)
s = silhouette_score(data1, cluster_labels, metric='euclidean', sample_size = 30)

print(f"V-Measure Score: {v}")
print(f"Silhouette Score: {s}")
```
