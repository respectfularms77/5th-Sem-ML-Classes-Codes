```python
import numpy as np
import pandas as pd

# The 2D array from the previous example
data = [
    [5, 12, 1, 9],
    [7, 2, 10, 4],
    [8, 11, 3, 6]
]
array_2d = np.array(data)

# Convert the NumPy array to a DataFrame
# Default indices (0, 1, 2) and columns (0, 1, 2, 3) are used
df = pd.DataFrame(array_2d)

print("Original DataFrame (3x4):\n", df)

# ----------------------------------------------------------------------

## 1. Using .loc (Label-Based Slicing)

# .loc[row_label_start : row_label_stop, col_label_start : col_label_stop]
# NOTE: .loc is **inclusive** for both start and stop labels.

print("\n--- .loc Slicing (Label-Based) ---")

# Slice the first two rows (labels 0 and 1) and columns 1 through 3 (labels 1, 2, 3)
# Row Labels: 0 up to 1 (inclusive)
# Column Labels: 1 up to 3 (inclusive)
loc_slice_1 = df.loc[0:1, 1:3]
print("Slice Rows 0-1, Cols 1-3 (Labels):\n", loc_slice_1)
# Resulting data: [[12, 1], [2, 10]]

# Slice just the last row (label 2) and all columns (using :)
loc_slice_2 = df.loc[2, :]
print("\nSlice Last Row (Label 2):\n", loc_slice_2)

# ----------------------------------------------------------------------

## 2. Using .iloc (Position-Based Slicing)

# .iloc[row_index_start : row_index_stop, col_index_start : col_index_stop]
# NOTE: .iloc is **exclusive** for the stop index.

print("\n--- .iloc Slicing (Position-Based) ---")

# Slice the first two rows (positions 0 and 1) and columns 0 and 1 (positions 0 and 1)
# Row Positions: 0 up to 2 (exclusive) -> Positions 0, 1
# Column Positions: 0 up to 2 (exclusive) -> Positions 0, 1
iloc_slice_1 = df.iloc[0:2, 0:2]
print("Slice Rows 0-1, Cols 0-1 (Positions):\n", iloc_slice_1)
# Resulting data: [[5, 12], [7, 2]]

# Slice the element at row position 2 and column position 3 (the last element)
iloc_slice_2 = df.iloc[2, 3]
print("\nSlice Element at Position [2, 3]:", iloc_slice_2)
```
